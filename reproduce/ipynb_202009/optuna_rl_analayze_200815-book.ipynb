```python
import optuna
import matplotlib.pyplot as plt
```

```python
study = optuna.load_study(study_name='rl_hyperparam_200810_100k', storage=r'sqlite:///D:\dev\pokeai\pokeai\experiment\gmm\rl\hyperparam_tuning_200810\optuna.db')
```

```python
study.get_trials()
```

```python
df = study.trials_dataframe(attrs=('number', 'value', 'params', 'state'))
```

```python
df
```

```python
study.trials[23]#best
```

```python
def make_chart_ax(ax, param_name, logx=False):
    if logx:
        ax.set_xscale("log")
    ax.scatter(df[:10]["params_"+param_name],df[:10].value, marker="x", label="first 10 trials", color="black")
    ax.scatter(df[10:]["params_"+param_name],df[10:].value, marker="o", label="last 14 trials", color="black")
    ax.set_xlabel(param_name)
    ax.set_ylabel("negative rate advantage")
    
```

```python
from pylab import rcParams

rcParams['figure.figsize'] = 10,5
```

```python
fig = plt.figure()
ax = fig.add_subplot(221)
ax.set_xlim(0.0,0.5)
make_chart_ax(ax, "epsilon")
ax = fig.add_subplot(222)
ax.set_xlim(3e-8,3e-5)
make_chart_ax(ax, "epsilon_decay",True)
ax = fig.add_subplot(223)
make_chart_ax(ax, "gamma")
ax = fig.add_subplot(224)
ax.set_xlim(3e-5,3e-1)
make_chart_ax(ax, "lr",True)
fig.tight_layout()
fig.savefig("fig/optuna_results.pdf")
fig.savefig("fig/optuna_results.png")
```
